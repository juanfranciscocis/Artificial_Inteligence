## Explique en qué consiste el cálculo lambda. ¿Cómo funciona la notación de tipos de funciones?

El cálculo lambda es un sistema que fue diseñado para investigar la definición de función, la noción de aplicación de funciones y la recursion. Este fue creado por Alonzo Church y Stephen Kleene en los 30´s. Church en el 36 logra calcular el Entscheidungsproblem usando lambda, entre otros colaboradores importantes se encuentra Albert Einstein. Hoy en dia el calculo lambda es considerado de los lenguajes de programación mas minimalistas y muchos lenguajes de programación modernos se basan en el calculo lambda. Lo interesante es que este consiste en una sola regla de reescrita, donde tenemos sustitución de variables y definición de funciones. El cálculo lambda es universal porque cualquier función computable puede ser expresada y evaluada a través de él. Por lo tanto, es equivalente a las máquinas de Turing.

En cuanto a como funciona la notacion del lenguaje lambda, tenemos varias formas de definir nuevas funciones combinando otras funciones, por ejemplo en Haskell tenemos la siguiente función:

**Definir si un entero es par o no:**

even :: Integral a => a -> Bool

even n = n ‘mod‘ 2 == 0”

Este ejemplo es interesante debido a que nos ayuda a comprender la notación de tipos de funciones, donde tenemos que el tipo de la función es a -> Bool, donde a es un tipo de dato que pertenece a la clase Integral, y el resultado de la función es un booleano. En cuanto a la definición de la función, tenemos que la función even recibe un entero y devuelve un booleano, donde el resultado de la función es el resultado de la comparación del residuo de la división del entero entre 2 con 0.

La notación de tipos de funciones se usa para expresar cómo los argumentos y resultados de una función se relacionan en términos de tipos de datos.

## ¿Qué es una expresión lambda? (Incluya ejemplos en Haskell, Python y C++).

Una expresión lambda es una funcion anonima que puede ser usada en cualquier lugar donde se requiera un objeto de tipo función. Estas funciones toman argumentos, realizan una operación y devuelven un valor. En lenguajes de programación como Python y C++ las usamos para crear funciones que no necesitan un nombre, por lo que se les conoce como funciones anónimas. Y Haskell es un lenguaje de programación funcional, por lo que todas las funciones son expresiones lambda.

#### Haskell
```
double x = x + x // Se define la funcion double, que recibe un argumento x y devuelve el resultado de la suma de x con x
```
**Ilustramos de mejor manera lo que hace la funcion double:**
```
double 3
= { aplicando double }
3 + 3
= { aplicando + }
6
```

#### Python
```
double = lambda x: x + x // Se define la funcion double, que recibe un argumento x y devuelve el resultado de la suma de x con x
```
**Ilustramos de mejor manera lo que hace la funcion double:**
```
double(3) = lambda 3: x + x
= { aplicando lambda }
3 + 3
= { aplicando + }
6
```

#### C++
```
auto doble = [](int x) { return x + x; }; // Se define la funcion double, que recibe un argumento x y devuelve el resultado de la suma de x con x
```
**Ilustramos de mejor manera lo que hace la funcion double:**
```
double resultado = doble(3);
doble(3) = [](int 3) { return x + x; };
= { aplicando lambda }
3 + 3
= { aplicando + }
6
```

En las matemáticas, la "list comprehension" es una notacion que nos ayuda a construir conjuntos a partir de conjuntos ya existentes, por ejemplo {x^3|xE{1..5}} produce el conjunto {1,8,27,64,125}. 

#### En Haskell, la "list comprehension" es una forma de definir listas, donde se puede especificar una lista por comprensión de la siguiente manera:
```
> [x^3 | x <- [1..5]]
[1,8,27,64,125]
```

#### En Python, la "list comprehension" es una forma de definir listas, donde se puede especificar una lista por comprensión de la siguiente manera:
```
import math
listComp = [int(math.pow(x,3)) for x in range(1,6)]
print(listComp)
[1,8,27,64,125]
```


## Explique los diferentes tipos de recursión que existen. (Incluya ejemplos distintos a los del libro). 

En Haskell, las funciones pueden ser definidas en terminos de otras funciones, pero tambien como en los lenguajes de programación que conocemos una función puede ser definida en terminos de esa misma función, a esto es lo que llamamos recursividad . Por ejemplo, la funcion `sum` puede ser definida como:

```
sum [] = 0
sum (x:xs) = x + sum xs
```
```
sum [1,2,3,4,5]
= { aplicando sum }
1 + sum [2,3,4,5]
= { aplicando sum }
1 + (2 + sum [3,4,5])
= { aplicando sum }
1 + (2 + (3 + sum [4,5]))
= { aplicando sum }
1 + (2 + (3 + (4 + sum [5])))
= { aplicando sum }
1 + (2 + (3 + (4 + (5 + sum []))))
= { aplicando sum }
1 + (2 + (3 + (4 + (5 + 0))))
= { aplicando + }
15
```
Como podemos ver hemos llamado a la función sum dentro de la misma función, esto es lo que llamamos recursividad.En este caso vimos un ejemplo de **recursividad en listas**.

Otros casos de recursividad son la **recursividad de multiples argumentos**, que es definida como la recursion en mas de un argumento, por ejemplo la funcion `MCD` puede ser definida como:

```
mcd a 0 = a
mcd a b = mcd b (a `mod` b)
```
```
mcd 12 8
= { aplicando mcd }
mcd 8 (12 `mod` 8)
= { aplicando mcd }
mcd 8 4
= { aplicando mcd }
mcd 4 (8 `mod` 4)
= { aplicando mcd }
mcd 4 0
= { aplicando mcd }
4
```

Otro caso común es la **recursividad multiple**, que es definida como la recursion en mas de una funcion, por ejemplo la funcion `potencia` puede ser definida como:

```
potencia :: Double -> Int -> Double
potencia x 0 = 1.0
potencia x n
    | n > 0 = x * potencia x (n-1) // N es positivo
    | n < 0 = 1.0 / x * potencia x (-n) // N es negativo
```
```
potencia 2 3
= { aplicando potencia }
2 * potencia 2 (3-1)
= { aplicando potencia }
2 * (2 * potencia 2 (2-1))
= { aplicando potencia }
2 * (2 * (2 * potencia 2 (1-1)))
= { aplicando potencia }
2 * (2 * (2 * 1.0))
= { aplicando * }
8.0
```

Finalmente tenemos la **recursividad mutua**, que es definida como la recursion entre dos o mas funciones, por ejemplo las funcion `turnoJuagador` y `turnoMaquina` pueden ser definidas como:

```
**En caso de que uno de los dos llegue a 0, gana**

//Simula el turno del jugador
turnoJugador :: Int -> String
turnoJugador 0 = "Gana Jugador"
turnoJugador n = turnoMaquina (n-1)

//Simula el turno de la maquina
turnoMaquina :: Int -> String
turnoMaquina 0 = "Gana Maquina"
turnoMaquina n = turnoJugador (n-1)
```
```
turnoJugador 3
= { aplicando turnoJugador }
turnoMaquina (3-1)
= { aplicando turnoMaquina }
turnoMaquina 2
= { aplicando turnoMaquina }
turnoJugador (2-1)
= { aplicando turnoJugador }
turnoJugador 1
= { aplicando turnoJugador }
turnoMaquina (1-1)
= { aplicando turnoMaquina }
turnoMaquina 0
= { aplicando turnoMaquina }
"Gana Maquina"
```
